```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("image.jpg")
cv2_imshow(image)
```

```python
rbg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2_imshow(rbg)
```

```python
gray_scale = cv2.cvtColor(rbg, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale)
```

```python
inverted_gray_scale = 255 - gray_scale
cv2_imshow(inverted_gray_scale )
```

```python
blur_image = cv2.GaussianBlur(inverted_gray_scale, (21,21), 0)
cv2_imshow(blur_image)
```

```python
inverted_blur = 255 - blur_image
cv2_imshow(inverted_blur)
```

```python
pencil_sketch = cv2.divide(gray_scale, inverted_blur, scale=255)
cv2_imshow(pencil_sketch)
```
